```python
from gtts import gTTS
from IPython.display import Audio

# Urdu script
urdu_text = "السلام علیکم! یہ اردو میں ٹیکسٹ ٹو اسپیچ کا ایک ڈیمو ہے۔"

# Convert to speech
tts = gTTS(text=urdu_text, lang='ur')
tts.save("urdu_audio.mp3")

# Play the audio
Audio("urdu_audio.mp3")
```

```python
with open('/kaggle/input/urduscript/urduscript.txt', 'r', encoding='utf-8') as f:
    urdu_text = f.read()

tts = gTTS(text=urdu_text, lang='ur')
tts.save("long_script_urdu.mp3")
Audio("long_script_urdu.mp3")
```

```python
from gtts import gTTS
from IPython.display import Audio
import gradio as gr

# ✅ Define Urdu TTS function
def urdu_tts(text):
    tts = gTTS(text=text, lang='ur')
    tts.save("urdu_voice.mp3")
    return "urdu_voice.mp3"

# ✅ Gradio Interface
app = gr.Interface(
    fn=urdu_tts,
    inputs=gr.Textbox(label="اردو متن یہاں درج کریں", lines=5, placeholder="ماحولیاتی تحفظ بہت ضروری ہے..."),
    outputs=gr.Audio(type="filepath", label="🔊 اردو آواز"),
    title="🗣️ اردو اسکرپٹ کو آواز میں تبدیل کریں",
    description="یہ ایپ آپ کے اردو متن کو آواز میں تبدیل کرے گی۔"
)

# ✅ Launch the app
app.launch()
```
